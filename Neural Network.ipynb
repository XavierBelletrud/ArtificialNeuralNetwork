### Descente de gradient pour les réseaux de neurone

La majorité des réseaux de neurone artificiels s'appuie sur l'algorithme de descente de gradient. Il est indispensable de comprendre les fondements de cet algorithme avant d'étudier les réseaux de neurone.
La descente de gradient est un algorithme d'optimisation permettant de trouver le minimum d'une fonction 

## Exercice 1

```python
import numpy as np
import matplotlib.pyplot as plt 
from tensorflow import keras

# Parameters (y = ax+b )
a = 0.6
b = 2

# Create Noisy data

x_data = np.linspace(-10, 10, num = 100000)   #np.linspace permet de créer 100000 elements compris entre -10 et 10
y_data = a * x_data + b + np.random.normal(size=100000)

plt.scatter(x_data[::500], y_data[::500])
plt.grid()
plt.show()
```

### Create the model with a single linear neuron 


```python
#Create the model with a single linear neuron 

model = keras.Sequential()
model.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
model.compile(loss='mse', optimizer ='adam')

# Display the model (only 2 parameters to optimize)
model.summary()

# Learn 
model.fit(x_data, y_data, epochs=5, verbose=1)
```

```python
# Predict (compute ) the output
y_predicted = model.predict(x_data)

# Display the result 

plt.scatter(x_data[::500], y_data[::500])
plt.plot(x_data, y_predicted, 'r', linewidth=4)
plt.grid()
plt.show()
```

```python
print(model.trainable_variables)
```

## Exercice 2

```python
import math



x_data = np.linspace(-10, 10, num = 1000)   #np.linspace permet de créer 100000 elements compris entre -10 et 10
y_data = 0.1 * x_data * np.cos(x_data) + 0.1 * np.random.normal(size=1000)

plt.scatter(x_data[:500], y_data[:500])
plt.grid()
plt.show()
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
model.add(keras.layers.Dense(units = 64, activation = 'relu'))
model.add(keras.layers.Dense(units = 64, activation = 'relu'))
model.add(keras.layers.Dense(units = 1, activation = 'linear'))
model.compile(loss='mse', optimizer="adam")

#loss est la perte ou l'erreur calculée sur la base des moindres carrés (loss='mse')

model.summary()
```

***x_data*** est l'entrée <br>
***y_data*** est la sortie désirée <br>
***epochs=5*** signifie que notre réseau sera entrainé 5 fois sur notre jeu de données <br>
***verbose = 1*** afiiche la progression dans la console <br>

```python
model.fit(x_data, y_data, epochs=500, verbose=1)
```

```python
# Compute the output
y_predicted = model.predict(x_data)

# Display the result
plt.scatter(x_data[::1], y_data[::1])
plt.plot(x_data, y_predicted, 'r', linewidth=4)
plt.grid()
plt.show()
```
